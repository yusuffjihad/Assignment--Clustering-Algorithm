```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, [3, 4]].values
```

```python
X
```

```python
from sklearn.cluster import SpectralClustering
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
# Generate sample data
blobs = make_blobs(n_samples=300, centers=4, random_state=42)
```

```python
# Apply Spectral Clustering
spectral_clustering = SpectralClustering(n_clusters=4, affinity='nearest_neighbors', random_state=42)
labels = spectral_clustering.fit_predict(X)
```

```python
# Visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', s=50, edgecolor='k')
plt.title('Spectral Clustering Result')
plt.show()
```
