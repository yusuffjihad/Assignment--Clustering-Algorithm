```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, [3, 4]].values
```

```python
X
```

```python
from sklearn.cluster import AffinityPropagation
from sklearn.datasets import make_blobs
```

```python
# Generate sample data
Makeblobs = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)
```

```python
# Apply Affinity Propagation
affinity_propagation = AffinityPropagation(damping=0.9)
labels = affinity_propagation.fit_predict(X)
```

```python
# Visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', s=50, edgecolor='k')
plt.title('Spectral Clustering Result')
plt.show()
```
